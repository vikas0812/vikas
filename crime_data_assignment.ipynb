```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crime=pd.read_csv('crime_data.csv')
crime
```

```python
#normalizing the data
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df=norm_func(crime.iloc[:,1:])
df
```

```python
#correlation
df.corr()
```

```python
import scipy
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
#generating dendrogram
z=linkage(df,method='ward',metric='euclidean')
plt.figure(figsize=(25,10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample_index')
plt.ylabel('distance')
dendrogram(z,leaf_rotation=90,leaf_font_size=15)
plt.show()
```

```python
z=linkage(df,method='complete',metric='euclidean')
plt.figure(figsize=(25,10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample_index')
plt.ylabel('distance')
dendrogram(z,leaf_rotation=90,leaf_font_size=10)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering=AgglomerativeClustering(n_clusters=3,linkage='ward',affinity='euclidean')
y_hc=clustering.fit_predict(df)
y_hc
```

```python
X=np.array(df)
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='blue')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='green')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='yellow')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=100,c='red')
```

```python
clustering=AgglomerativeClustering(n_clusters=3,linkage='average',affinity='euclidean')
y_hc=clustering.fit_predict(df)
y_hc
X=np.array(df)
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='blue')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='green')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='yellow')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=100,c='red')
```
