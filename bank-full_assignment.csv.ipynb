```python
y=df.iloc[:,16:17]
y
```

```python
x=df.iloc[:,0:16]
x
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_x=LabelEncoder()
x['job']=labelencoder_x.fit_transform(x['job'])
x['marital']=labelencoder_x.fit_transform(x['marital'])
x['education']=labelencoder_x.fit_transform(x['education'])
x['default']=labelencoder_x.fit_transform(x['default'])
x['housing']=labelencoder_x.fit_transform(x['housing'])
x['loan']=labelencoder_x.fit_transform(x['loan'])
x['contact']=labelencoder_x.fit_transform(x['contact'])
x['month']=labelencoder_x.fit_transform(x['month'])
x['poutcome']=labelencoder_x.fit_transform(x['poutcome'])
x
```

```python
y['y']=labelencoder_x.fit_transform(y['y'])
y
```

```python
A=pd.concat([x,y],axis=1)
A
```

```python
A.corr()['y']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logisticregression=LogisticRegression()
logisticregression.fit(train_x,train_y)
```

```python
y_pred=logisticregression.predict(test_x)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,test_y)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,test_y)
```
